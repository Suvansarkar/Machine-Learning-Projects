```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
class LogitRegression:
    def __init__(self, alpha, iterations):
        self.alpha = alpha # Learning Rate
        self.iterations = iterations

    def fit(self, X, Y):
        self.m, self.n = X.shape
        self.w = np.random.rand(self.n)
        # self.w = np.zeros(self.n)
        self.b = 0
        for _ in range(self.iterations):
            fx = 1 / (1 + np.exp(-(np.dot(X, self.w) + self.b)))
            dw = (1 / self.m) * np.dot(X.T, (fx - Y))
            db = (1 / self.m) * np.sum(fx - Y)
            self.w = self.w - self.alpha * dw
            self.b = self.b - self.alpha * db
        print("Bias: ", self.b)
        print("Weights: ", self.w)
        print("Train accuracy: ", self.check_accuracy(Y, self.predict(X)))

    def predict(self, X_test):
        Y_pred = 1 / (1 + np.exp( - (X_test.dot(self.w) + self.b ) )) 
        Y_pred = np.where( Y_pred > 0.5, 1, 0)
        return Y_pred
    
    @staticmethod
    def check_accuracy(Y, Y_pred):
        count = 0
        return float(np.mean(Y == Y_pred)*100)
    
```

```python
X = data.iloc[:, 0:60]
Y = data[60]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42, shuffle=True, stratify=Y)
```

```python
X_train
```

```python
model = LogitRegression(1.3, 100)
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
print("Test Accuracy: ", LogitRegression.check_accuracy(Y_test, Y_pred))
```
